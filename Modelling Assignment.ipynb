```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
import statsmodels.formula.api as sm


```

```python
#1)i)	Read the dataset, print last 3 observations 

df = pd.read_csv("Demand.csv",index_col=0)
df.tail(3)

```

```python
#1)ii)Find the data types of each column
df.dtypes

```

```python
#2)â€¢	Convert the data type of â€˜Mapped_Item_Codeâ€™ field to object
#	Hint: use .astype(str) for the data type conversion
df["Mapped_Item_Code"]=df["Mapped_Item_Code"].astype(str)
df["Mapped_Item_Code"]
```

```python
#3)â€¢	Convert the field â€˜Transaction_Dateâ€™ to datetime (Date) column
#	Hint: use pd.to_datetime for conversion
df["Transaction_Date"]=pd.to_datetime(df["Transaction_Date"])
df["Transaction_Date"]
```

```python
#4)â€¢	Extract only those records whose transaction date is greater than or equal to â€˜2017-01-01â€™ and proceed with below operations
#	Hint: use .loc operator
df.loc[df['Transaction_Date'] >= "2017-01-01"] 
```

```python
#5)â€¢	Check the changed data types of each column and also see how many unique sales types are available in â€˜Mapped_Sales_Typeâ€™ column
print(df.dtypes)

df["Mapped_Sales_Type"].nunique()
```

```python
#6)â€¢	Extract only those records whose â€˜Avg_Discount_Percent_On_Discounted_Itemsâ€™ is lesser than 1.0 
#and do a sense check if the resulting data frame contains any values in the same column that are greater than or equal to 1.0
df.loc[df['Avg_Discount_Percent_On_Discounted_Items'] < 1.0]
```

```python
#7)â€¢	Create the below dictionary (dict_agg) and pass it as an argument for the aggregate function in order to roll up the
#results grouped by following order: 'City', 'Mapped_Sales_Type', 'Mapped_Item_Code', 'Transaction_Date'
#	dict_aggr = {'Quantity_Sold':'sum', 'Median_Price':'median', 'Effective_Price':'median'}
#	sample syntax: df.groupby(<groupby list>).agg(dictionary)
dict_aggr={'Quantity_Sold':'sum', 'Median_Price':'median', 'Effective_Price':'median'}
df.groupby(['City','Mapped_Sales_Type','Mapped_Item_Code','Transaction_Date']).agg(dict_aggr)
```

```python
#8)â€¢	Print first 5 observations where Effective_Price is not null
print(df["Effective_Price"].isna().sum())
df["Effective_Price"].dropna()
df["Effective_Price"].head()
```

```python
#9)â€¢	Print first 10 observations where â€˜Cityâ€™ is â€˜Hyderabadâ€™ AND â€˜Mapped_Sales_Typeâ€™ is â€˜Deliveryâ€™
#	Hint: df.loc[(<condition 1>) & (<condition 2>), :].head(10)

df.loc[(df['City']=='Hyderabad') & (df['Mapped_Sales_Type']=='Delivery')].head(10)
```

```python
#10)â€¢	Print first 3 observations where Percent_Quantity_With_Discount is not equal to zero
#	Hint: use != operator

df.loc[df["Percent_Quantity_With_Discount"]!=0].head(3)
```

```python
#11)â€¢	Create a new column â€˜Rangeâ€™ and populate it as 1 if Effective_Price is greater than 500, else 0.5
#	Hint: df[â€˜New columnâ€™] = np.where(<condition>, if true, if false)
df['Range']=np.where((df["Effective_Price"]>500),1,0.5)
df[['Effective_Price','Range']]
```

```python
#12)â€¢	Split the dataset into X and Y based on below list:
# x = 'Mapped_Item_Code', 'Median_Price', 'Average_Tax_Percent',  'Effective_Price',
# 'Percent_Quantity_With_Discount',       'Avg_Discount_Percent_On_Discounted_Items','Car_sales', 'F_and_B_Inflation', 
# 'F_and_B_Inflation_Adjusted_Effective_Price',       'F_and_B_Inflation_Adjusted_Median_Price'.         
# And  y = â€˜Quantity_Soldâ€™

X=df[['Mapped_Item_Code', 'Median_Price', 'Average_Tax_Percent',  'Effective_Price','Percent_Quantity_With_Discount',
      'Avg_Discount_Percent_On_Discounted_Items','Car_sales', 'F_and_B_Inflation',
     'F_and_B_Inflation_Adjusted_Effective_Price','F_and_B_Inflation_Adjusted_Median_Price']].fillna(0)

Y=df['Quantity_Sold']

```

```python
#13)â€¢	Run the Linear Regression model on the above x and y features. 
#Feel free to explore the modeling parameters and come up with a high R2  
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=123)
```

```python
lr = LinearRegression()
lr.fit(X_train,Y_train)
```

```python
train_prediction=lr.predict(X_train)
```

```python
metrics.mean_squared_error(Y_train,train_prediction)
```

```python
metrics.mean_absolute_error(Y_train,train_prediction)
```

```python
test_prediction=lr.predict(X_test)
```

```python
metrics.mean_squared_error(Y_test,test_prediction)
```

```python
metrics.mean_absolute_error(Y_test,test_prediction)
```

```python
metrics.r2_score(Y_test,test_prediction)
```
