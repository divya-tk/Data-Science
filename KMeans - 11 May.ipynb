```python
import pandas as pd
import numpy as np
from matplotlib import pyplot
from sklearn import cluster
```

KMeans   --> Unsupervised ---> if we don't know the o/p variable we ca use this to get similar data on particular clusters

```python

geyser_data = np.genfromtxt(fname='geyser.csv',delimiter=',',dtype=float)
```

```python
type(geyser_data)
```

```python
geyser_data[:5]
```

```python
for eruption_time,waiting_time in geyser_data:
    pyplot.scatter(eruption_time,waiting_time,c='b')
    
pyplot.title('Geyser')
pyplot.xlabel('Eruption')
pyplot.ylabel('waiting time')
pyplot.show()
```

```python
k = 2
kmeans = cluster.KMeans(n_clusters=k)
kmeans.fit(geyser_data)
```

```python
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
labels = kmeans.predict(geyser_data)   #splitting as 2 clusters 0 and 1
print(labels)
```

```python
for each in range(k):
    selected_data = geyser_data[np.where(labels==each)]
# print(selected_data)              #indexing of array
    pyplot.plot(selected_data[:,0],selected_data[:,1],'o')
    lines = pyplot.plot(centroids[each,0] ,centroids[each,1],'x')
    pyplot.setp(lines,markersize=15.0,markeredgewidth=2.0)
    
pyplot.title('k-Means Results')
pyplot.xlabel('Eruption Time')
pyplot.ylabel('Waiting Time')
pyplot.show()
```
